```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
#load pre-trained model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
#input text to summarize
text = """
Virat Kohli is an accomplished Indian cricketer, known for his powerful batting, especially in shorter formats of the game, and has captained the Indian national team to significant victories, including the 2008 Under-19 World Cup. A prolific scorer, Kohli's career is marked by his energetic on-field demeanor, though it can sometimes appear irritable, and he is a globally recognized figure in the sport.
"""
```

```python
#tokenize the input text
inputs = tokenizer.encode("Summury: "+text, return_tensors="pt", max_length=1024, truncation=True)
```

```python
#Generate the Summary:
summary_ids = model.generate(inputs, max_length=50, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
```
